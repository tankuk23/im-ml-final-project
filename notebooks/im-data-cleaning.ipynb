```python
def removeColours(df):

    btmpf = df[df.blueWins == 1]
    rtmpf = df[df.redWins == 1]
    
    newcols = {}
    for c in list(btmpf.columns):
        if "blue" in c:
            newcols[c] = c.replace("blue", "winner")
        elif "red" in c:
            newcols[c] = c.replace("red", "loser")
    btmpf.rename(columns=newcols, inplace=True, copy=True)
    btmpf['winnerColour'] = "blue"
    btmpf.drop(['winnerWins', 'loserWins'],axis=1)

    newcols = {}
    for c in list(rtmpf.columns):
        if "blue" in c:
            newcols[c] = c.replace("blue", "loser")
        elif "red" in c:
            newcols[c] = c.replace("red", "winner")
    rtmpf.rename(columns=newcols, inplace=True)
    rtmpf['winnerColour'] = "red"
    rtmpf.drop(['winnerWins', 'loserWins'],axis=1)

    return pd.concat([btmpf,rtmpf], ignore_index=True)
```

```python
cf = removeColours(df)
```

```python
cf.head()
```

```python
mf = removeColours(mf)
```

```python
gmf = removeColours(gmf)
```

```python
cf.to_csv("../data/colourless_challenger.csv")
```
